## Item-Based Collaborative Filtering

```python
userRatings = ratings.pivot_table(index='userId',columns='title',values='rating')
userRatings.head()
```

```python
corrMatrix = userRatings.corr()
corrMatrix.head()
```

```python
corrMatrix = userRatings.corr(method='pearson', min_periods=100)
corrMatrix.head()
```

```python
myRatings = userRatings.loc[3].dropna()
myRatings
```

```python
simCandidates = pd.Series()
for i in range(0,len(myRatings.index)):
    print("Adding sims for "+myRatings.index[i]+"...")
    sims = corrMatrix[myRatings.index[i]].dropna()
    sims = sims.map(lambda x:x*myRatings[i])
    simCandidates = simCandidates.append(sims)
print("sorting...")
simCandidates.sort_values(inplace=True,ascending =False)
simCandidates.head(10)
```

```python
simCandidates = simCandidates.groupby(simCandidates.index).sum()
```

```python
simCandidates.sort_values(inplace=True,ascending=False)
simCandidates.head(10)
```

```python
remove_list =[]
for i in simCandidates.index:
    for j in myRatings.index:
        if i==j:
            remove_list.append(i)
remove_list
```

```python
simCandidates[remove_list]
```

```python
myRatings['Shawshank Redemption, The (1994)']
```

```python
simCandidates.drop(remove_list)
```

```python
filtered_results = simCandidates.drop(remove_list)
filtered_results
```

```python
merged_results = pd.merge(movies_df.set_index('title'),pd.DataFrame(filtered_results,columns=['SimilarityScore']),how='outer', left_index=True, right_index=True)
```

```python
def search(searchString):
    filter = merged_results.index.str.contains(searchString)
    return merged_results[filter].dropna().sort_values(by=['SimilarityScore'],ascending=False)
```
